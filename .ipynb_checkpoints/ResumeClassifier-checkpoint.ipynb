```python
import os
import re
import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from PyPDF2 import PdfReader

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report
)

#please run below commands it`s your first time
# nltk.download('stopwords')
# nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|\S+@\S+", " ", text)
    text = re.sub(r"[^a-zA-Z]", " ", text)
    text = re.sub(r"\s+", " ", text)

    words = [
        lemmatizer.lemmatize(word)
        for word in text.split()
        if word not in stop_words
    ]

    return " ".join(words)
```

```python
# CSV must contain: category, text
df = pd.read_csv("Dataset.csv")

df = df[['Category', 'Text']].dropna()
df['clean_text'] = df['Text'].apply(clean_text)

print("Dataset loaded:", df.shape)
```

```python
encoder = LabelEncoder()
df['label'] = encoder.fit_transform(df['Category'])

print("Classes:", encoder.classes_)

```

```python
tfidf = TfidfVectorizer(
    ngram_range=(1, 2),
    max_features=1500
)

X = tfidf.fit_transform(df['clean_text'])
y = df['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,
    random_state=42,
    stratify=y
)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
lr_base = LogisticRegression(
    max_iter=1000,
    random_state=42,
    solver='liblinear'
)

lr=OneVsRestClassifier(lr_base)
lr.fit(X_train, y_train)

# print("Logistic Regression model trained successfully using OneVsRestClassifier.")
```

```python
def evaluate_model(model, X_test, y_test, name, encoder):
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro')
    recall = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')

    print(f"\n{name} Performance")
    print("-" * 40)
    print("Accuracy :", accuracy)
    print("Precision:", precision)
    print("Recall   :", recall)
    print("F1-Score :", f1)
    print("\nClassification Report:\n")
    print(classification_report(
        y_test,
        y_pred,
        target_names=encoder.classes_
    ))

    return accuracy, precision, recall, f1



knn_metrics=evaluate_model(knn, X_test, y_test, "KNN", encoder)
lr_metrics=evaluate_model(lr, X_test, y_test, "Logistic Regression", encoder)

```

```python
# Metrics
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']

knn_values = list(knn_metrics)
lr_values  = list(lr_metrics)

x = range(len(metrics))

plt.figure(figsize=(8, 5))
plt.bar(x, knn_values, width=0.35, label='KNN')
plt.bar([i + 0.35 for i in x], lr_values, width=0.35, label='Logistic Regression')

plt.xticks([i + 0.17 for i in x], metrics)
plt.ylabel("Score")
plt.title("Model Performance Comparison")
plt.legend()
plt.ylim(0, 1)

plt.show()
```

```python
joblib.dump(knn, "knn_model.pkl")
joblib.dump(lr, "logistic_model.pkl")
joblib.dump(tfidf, "tfidf_vectorizer.pkl")
joblib.dump(encoder, "label_encoder.pkl")
```

```python
def extract_text_from_pdf(pdf_path):
    text = ""
    reader = PdfReader(pdf_path)

    for page in reader.pages:
        page_text = page.extract_text()
        if page_text:
            text += page_text + " "

    return text
```

```python
def predict_pdf_files(folder_path, model):
    results = []

    for file in os.listdir(folder_path):
        if file.lower().endswith(".pdf"):
            pdf_path = os.path.join(folder_path, file)

            raw_text = extract_text_from_pdf(pdf_path)
            cleaned_text = clean_text(raw_text)

            vector = tfidf.transform([cleaned_text])
            pred = model.predict(vector)[0]
            category = encoder.inverse_transform([pred])[0]

            results.append((file, category))

    return results
```

```python
print("\nSelect Model:")
print("1 → KNN")
print("2 → Logistic Regression")

choice = input("Enter choice (1 or 2): ")

selected_model = knn if choice == "1" else lr

resume_folder = "input_resumes"   # folder containing PDF resumes

predictions = predict_pdf_files(resume_folder, selected_model)

print("\nResume Classification Results:\n")
for file, category in predictions:
    print(f"{file}  →  {category}")
```

```python
def match_resumes_with_jd(resume_folder, job_description, tfidf):
    jd_clean = clean_text(job_description)
    jd_vector = tfidf.transform([jd_clean])

    scores = []

    for file in os.listdir(resume_folder):
        if file.lower().endswith(".pdf"):
            pdf_path = os.path.join(resume_folder, file)

            resume_text = extract_text_from_pdf(pdf_path)
            resume_clean = clean_text(resume_text)
            resume_vector = tfidf.transform([resume_clean])

            similarity = cosine_similarity(jd_vector, resume_vector)[0][0]
            scores.append((file, similarity))

    # Sort by best match
    scores.sort(key=lambda x: x[1], reverse=True)
    return scores
```

```python
print("\n--- Job Description to Resume Matching ---")
job_description = input("Enter Job Description: ")

matches = match_resumes_with_jd(
    resume_folder="input_resumes",
    job_description=job_description,
    tfidf=tfidf
)

print("\nResume Match Scores:\n")
for file, score in matches:
    print(f"{file}  →  Similarity: {score:.4f}")
```
